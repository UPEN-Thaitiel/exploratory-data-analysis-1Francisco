```python
years = [2010, 2011, 2012, 2013]
sales = [100, 120, 140, 160]
```

```python
plt.plot(years, sales, label= "blue", marker="o")
plt.title("Sales  over the years")
plt.xlabel("Years")
plt.ylabel("Sales")
```
